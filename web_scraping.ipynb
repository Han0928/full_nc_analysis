```python
import numpy as np
import matplotlib.pyplot as plt

# Generate some example data for x, y, u, v (wind speed components)
x = np.linspace(0, 10, 20)
y = np.linspace(0, 10, 20)
X, Y = np.meshgrid(x, y)
u = np.sin(X)
v = np.cos(Y)

# Calculate wind speed magnitude and direction
speed = np.sqrt(u**2 + v**2)
direction = np.arctan2(v, u)

# Create the quiver plot
fig, ax = plt.subplots()
q = ax.quiver(X, Y, u, v, speed, cmap='cool')
ax.quiverkey(q, X=0.9, Y=1.05, U=10, label='Wind speed (m/s)', labelpos='E')

# Add a colorbar and axis labels
cb = fig.colorbar(q)
cb.set_label('Wind speed (m/s)')
ax.set_xlabel('x')
ax.set_ylabel('y')

# Show the plot
plt.show()
```
